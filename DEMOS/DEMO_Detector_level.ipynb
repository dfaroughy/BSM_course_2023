`run_01/tag_1_delphes_events.lhco`

in the `Events` folder at the end of the run.

The name comes from "Les Houches Collider Output". The files contain events that have beed showered and hadornized with `Pythia`, passed thhrough a fast detector simulator `Delphes` and jet clustering `FastJet`. Each event in the file is composed by a handful of *high-level reconstructed objects* and other atributes like elctric charge. Most importnatly, these objects come with an identification label that reflects the capabilities of the ATLAS and CMS detectors to uncover the identity of the underlying fundamental particles produced in the hard collision.

Types of objects that are reconstructed at the LHC:
- **0: isolated photons**. Proxy for a prompt photon.
- **1: isolated electrons**. Proxy for a prompt electron.
- **2: isolated muons**. Proxy for a prompt muon.
- **3: $\tau$-tagged jets**, defined by a jet clustering algorithm. Proxy for a hadronically decaying $tau$-lepton.
- **4: jets**, also defined by a jet clustering algorithm. Proxy for a hard quark or gluon. For narrow jets there are two  sub-cathegories:
    - **$b$-jets**. Proxy for a heavy $b$-quark.
    - **light-jets**. Proxy for a light quark or gluon.
- **6: MET** or missing transverse energy. Proxy for a "stable" neutral particle, e.g. neutrinos.



Here's a breakdown of the columns for each object in an LHCO event:

1. `typ`: This is an integer specifying the type of object as described above.

2. `pt`: Transverse momentum of the object (in GeV).

3. `eta`: Pseudorapidity of the object.

4. `phi`: Azimuthal angle (in radians) of the object.

5. `jmas`: Invariant mass of the object. This is most relevant for jets. For other objects like electrons, muons, and photons, this is usually set to 0. 

6. `ntrk`: Number of tracks associated with the object. Correpsonds to the electric charge for leptons. Ignore for jets.

7. `btag`: b-tagging information. This variable indicates if the object (jet) is identified as coming from a b-quark. The value is 0 and 1. Best to ignore for non-jet objects

8. `had/em`: Hadronic to electromagnetic calorimeter energy fraction. To my knowledge this information is not reliable for studies.

9. `dum1`, `dum2`: dummy columns.

## Lepton / Photon isolation

An **isolated lepton** is a lepton canididate that passes the following *isolation requirement*:

$$\frac{\sum_{x\in R_{iso}} p_T(x) }{ p_T(\ell)}\ <\ I_{iso}$$

where $x\in R_iso$ correspond to all particles at a distance $R_{iso}$ (dubbed the isolation radius) of the candidate lepton $\ell$.
The procedure has two free parameters: the isolation cone radius $R_{iso}$ and the energy ratio threshold $I_{iso}. These are usually fixed depending on the typoe of search.

The fast detector simulator `Delphes` has a module that performs the isolation. The parameters can be fixed in the delphes_card_CMS.dat or delphes_card_ATLAS.dat:

    ################
    # Muon isolation
    ################

    module Isolation MuonIsolation {
    set CandidateInputArray MuonEfficiency/muons
    set IsolationInputArray EFlowFilter/eflow

    set OutputArray muons

    set DeltaRMax 0.5   <---- R_iso

    set PTMin 0.5

    set PTRatioMax 0.25   <---- I_iso
    }

```python

from BSMhighPT2023.isolation import isolated_object

N_muons = 0
N_isolated_muons = 0 

for event in sample:

    #...select muons in central region:

    muon_mask = abs(event['pid']) == 13
    muons = event[muon_mask]
    hadrons = event[~muon_mask] 

    #...apply isolation requirement:

    if len(muons) > 0:

        N_muons += len(muons)
        isolated_muons, hadrons = isolated_object(isolated_candidates=muons, hadrons=hadrons, R_iso=0.5, threshold=0.12)

        N_isolated_muons += len(isolated_muons)
            
print('muon isolation efficiency = ', N_isolated_muons / N_muons)
```

# Jet clustering

Once the isolated leptons are extracted from the event, the next step is to cluster the remaining hadornic activity into jets.


Jets are one of the most populous objects apearing at hadron colliders. Physically, they correspond to collimated sprays of hadrons that are a conseuqnece of QCD. IN practice, jets are reconstructed with a jet-finding or clustering algorithm. The most used algorithm now a days (e.g. at the LHC) are the so-called *sequential recombination algorithms, like , e.g. the Anti-k_T algorithm. These class of algorithms are based on a distance between pairs of 4-vectors and require radius as input parameter $R$ dictating how large are the jets being looked for.

The main software for clustering jets is called `FastJet` (C++). In our setup, the jet clustering parameters can be modified in the `Delphes`:

    #############
    ## Jet finder
    #############
    #
    #module FastJetFinder FastJetFinder {
    ##  set InputArray Calorimeter/towers
    #  set InputArray EFlowMerger/eflow
    #
    #  set OutputArray jets
    #
    #  # algorithm: 1 CDFJetClu, 2 MidPoint, 3 SIScone, 4 kt, 5 Cambridge/Aachen, 6 antikt

    #  set JetAlgorithm 6             <---  set algorithm
    #  set ParameterR 0.4             <---  set Radius size
    #  set JetPTMin 20.0              <---  only keeps jets that have pt larger than this value
    #}


In the same card there are also modules for jet tagging that are important for BSM studies. The Delphes taggers will look for a truth level b-quark or hadronic tau in the event and will randomly tag the asssociated jet as a b-jet or tau-jet dependeing on provided tagging efficiencies and mis-identifications rates. These efficiencies are inputs that vary from search to search. One can fix these in tghe Delphes card here:   

- B-tagging Module:

        ############
        ## b-tagging
        ############
        #
        #module BTagging BTagging {
        #  set JetInputArray JetEnergyScale/jets
        #
        #  set BitNumber 0
        #
        #  # add EfficiencyFormula {abs(PDG code)} {efficiency formula as a function of eta and pt}
        #  # PDG code = the highest PDG code of a quark or gluon inside DeltaR cone around jet axis
        #  # gluon's PDG code has the lowest priority
        #
        #  # based on arXiv:1211.4462
        #  
        #  # default efficiency formula (misidentification rate)
        #  add EfficiencyFormula {0} {0.01+0.000038*pt}                             <----- prob that a jet from g, q is (wrongly) tag as a b-jet
        #
        #  # efficiency formula for c-jets (misidentification rate)
        #  add EfficiencyFormula {4} {0.25*tanh(0.018*pt)*(1/(1+ 0.0013*pt))}       <----- prob that a jet from charm is (wrongly) tag as a b-jet
        #
        #
        #  # efficiency formula for b-jets
        #  add EfficiencyFormula {5} {0.85*tanh(0.0025*pt)*(25.0/(1+0.063*pt))}     <----- prob that a jet from botton is (correctly) tag as a b-jet

## jet clustering in python

In python there is a relatively new library called `fastjet`. It is pretty much an interface between  `FastJet`(C++) and `NumPy` based on something called `akward`. It seems actively mantained on github, but that may change...

**Beware:  If you wish to perform jet clustering "professionally" it is better to use FsstJet directly in C++. It is much faster, actively mantained and has passed the test of time.** 

Following `FastJet` on has to first define the jet algorithm and fix the radius size $R$ of the seeked jets:  

```python
import fastjet

jet_def = fastjet.JetDefinition(fastjet.antikt_algorithm, 0.4)
jet_def
```

You can use a wrapper function `get_jets` that performs the clustering and extracts the jets.

Following the LHCO numberign scheme for recontructed objectsm, we assign "4" for the `pid` field of the jet.

```python

from BSMhighPT2023.isolation import isolated_object
from BSMhighPT2023.fastjet import get_jets

jets = []
isolated_leptons = []

for event in sample:

    #...1) get isolated leptons (we do not want to cluster isolated letpons):

    lepton_mask = (abs(event['pid']) == 11) | (abs(event['pid']) == 13)
    leptons = event[lepton_mask]
    hadrons = event[~lepton_mask] 

    if len(leptons) > 0:
        leptons, hadrons = isolated_object(isolated_candidates=leptons, hadrons=hadrons, R_iso=0.5, threshold=0.12)
        pt_mask = leptons['pt'] > 20.0
        leptons = leptons[pt_mask]
    else: 
        hadrons = event    

    isolated_leptons.append(leptons)

    #...2) cluster remaining hadrons into jets here:

    clustered_jets = get_jets(hadrons, jet_def, pt_min=20.0)
    jets.append(clustered_jets)


#...inspect jets in first event:

print(jets[0])
```
